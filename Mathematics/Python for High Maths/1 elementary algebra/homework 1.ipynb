#### Парабола № 1

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
# функция для параболы
x = np.linspace(0,6,1000)
y = -2.125*(x**2) + 12.5*x - 8.375
# точки, через которые должна проходить парабола:
x1, y1 = 1, 2
x2, y2 = 3, 10
x3, y3 = 5, 1
```

```python
plt.figure(figsize=(10,5))
# точки
plt.plot(x1, y1, c='b', marker = 'o')
plt.plot(x2, y2, c='b', marker = 'o')
plt.plot(x3, y3, c='b', marker = 'o')
# парабола
plt.plot(x, y, c='g')
plt.title("Парабола")
plt.xlabel('x')
plt.ylabel('y')

plt.grid()
plt.show()
```
