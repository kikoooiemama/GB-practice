```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import warnings
warnings.filterwarnings('ignore')
```

```python
def draw_vector(x1, x2, x3):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    X, Y, Z, U, V, W = 0, 0, 0, x1, x2, x3
    ax.quiver(X, Y, Z, U, V, W)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.set_xlim([-1-np.abs(x1),np.abs(x1)+1])
    ax.set_ylim([-1-np.abs(x2),np.abs(x2)+1])
    ax.set_zlim([-1-np.abs(x3),np.abs(x3)+1])
    plt.show()
```

#### № 4. Нарисуйте трехмерные графики поверхностей 2-го порядка

<u>Эллипсоид</u>

```python
def get_ellipsoid_by_spherical_cs(a,b,c,x0,y0,z0):
    # Углы:
    phi = np.linspace(0, 2 * np.pi, 100) # азимутальный угол, u
    theta  = np.linspace(0, np.pi, 100) # полярный угол, v
    
    # Параметрическое уравнение эллипсоида:
    x = a * np.outer(np.cos(phi), np.sin(theta)) - x0
    y = b * np.outer(np.sin(phi), np.sin(theta)) - y0
    z = c * np.outer(np.ones_like(phi), np.cos(theta)) - z0
    
    return x,y,z
```

```python
def get_ellipsoid_by_cartesian_cs(a,b,c,x0,y0,z0):
    X1 = np.linspace(a, -a, 1000, endpoint=True)
    Y1 = np.linspace(b, -b, 1000, endpoint=True)
    X1 = X1 - x0
    Y1 = Y1 - y0
    X1, Y1 = np.meshgrid(X1, Y1)
    Z1 = c * np.sqrt(1 - np.power(X1,2)/np.power(a,2) - np.power(Y1,2)/np.power(b,2)) - z0
    
    X2 = -X1
    Y2 = -Y1
    Z2 = -c * np.sqrt(1 - np.power(X2,2)/np.power(a,2) - np.power(Y2,2)/np.power(b,2)) - z0
    
    X = np.concatenate((X1, X2))
    Y = np.concatenate((Y1, Y2))
    Z = np.concatenate((Z1, Z2))
    
    return X, Y, Z
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = figure()
ax = Axes3D(fig)
fig.set_size_inches(14, 8, forward=True)

a, b, c = 10, 20, 10
x0, y0, z0 = 0, 0, 100 # смещение по осям

x,y,z = get_ellipsoid_by_spherical_cs(a,b,c,x0,y0,z0)

# График:
ax.plot_wireframe(x, y, z,  rstride=4, cstride=4, color='steelblue')
    
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("Эллипсоид")
    
plt.show()
```

```python
fig = figure()
ax = Axes3D(fig)
fig.set_size_inches(14, 8, forward=True)
a,b,c = 10, 15, 15
x0,y0,z0 = 0,0,100
X, Y, Z = get_ellipsoid_by_cartesian_cs(a,b,c,x0,y0,z0)
ax.plot_wireframe(X, Y, Z, color = 'steelblue')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("Эллипсоид")

plt.show()
```

```python
def polar_to_cartesian(r, phi):
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return x, y
```

### № 9

Решите систему уравнений:
$ \begin{cases}
    y = x^2 - 1 \\
    \exp{(x)} + x(1-y)=1
  \end{cases}$

```python
%matplotlib inline
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

from pylab import rcParams
rcParams["figure.figsize"] = 8, 8
 
x = np.linspace(-5, 10, 201)
plt.plot(x, (1 - (1-np.exp(x))/x))
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2,25) 
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x)+x*(1-y)-1)

#print(equations ([-1, 1]))
result = []
result.append(fsolve(equations, (-2, 2)))
result.append(fsolve(equations, (2.5, 5)))
result.append(fsolve(equations, (4,15)))
       
for i, res in enumerate(result, 1):
    print(f"{i} решение: ({round(res[0],3)}, {round(res[1],3)})")
```
